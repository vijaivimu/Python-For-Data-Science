```python
from typing import List

def square_elements(numbers: List[int]):
    """
    This function takes a list of integers and returns a new list
    where each integer is squared.
    """
    for num in numbers:
        print(num)

# Example usage
input_numbers = [1, 2, "Test", 4]
output_numbers = square_elements(input_numbers)
print(output_numbers)  # Output: [1, 4, 9, 16]
```
