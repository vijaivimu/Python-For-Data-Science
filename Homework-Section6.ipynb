```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('section6hw.csv', encoding='latin1')
```

```python
data.head()
```

```python
filtered_genre=['action','comedy','adventure','animation','drama']
data[data['Genre'].isin(filtered_genre)]
```

```python
data.columns
```

```python
sns.set(style='darkgrid',palette='dark:black')
sns.boxplot(data=data[data['Genre'].isin(filtered_genre)],x='Genre',y='Gross % US',color='grey')

# Overlay the data points on top of the boxplot
sns.stripplot(data=data[data['Genre'].isin(filtered_genre)], 
              x='Genre', y='Gross % US', 
              color='black', size=6, jitter=False,hue='Studio')
```
