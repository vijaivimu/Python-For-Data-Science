```python
mov['Genre'].isin(['action', 'adventure', 'animation', 'comedy', 'drama'])
```

```python
# Filter the dataframe by genre
mov2 = mov[mov['Genre'].isin(['action', 'adventure', 'animation', 'comedy', 'drama'])]
```

```python
mov2['Studio'].isin(['Buena Vista Studios','Fox', 'Paramount Pictures', 'Sony', 'Universal','WB'])
```

```python
# Filter the dataframe by studio
mov3 = mov2[mov2['Studio'].isin(['Buena Vista Studios','Fox', 'Paramount Pictures', 'Sony', 'Universal','WB'])]
```

```python

mov3['Genre'] = mov3['Genre'].cat.remove_unused_categories()
mov3['Studio'] = mov3['Studio'].cat.remove_unused_categories()
# Check how the filters worked
print (mov3.Genre.unique())
print (mov3.Studio.unique())
print (len(mov3))
```

```python
# Define the style
sns.set(style="darkgrid", palette="muted", color_codes=True)

# Plot the boxsplots
ax = sns.boxplot(data=mov3, x='Genre', y='Gross % US', orient='v', color='lightgray', showfliers=False)
plt.setp(ax.artists, alpha=0.5)

# Add in points to show each observation
sns.stripplot(x='Genre', y='Gross % US', data=mov3, jitter=True, size=6, linewidth=0, hue = 'Studio', alpha=0.7)

#ax.axes.set_title('_',fontsize=30)
#ax.set_xlabel('_',fontsize=20)
#ax.set_ylabel('_',fontsize=20)

# Define where to place the legend
#ax.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```
